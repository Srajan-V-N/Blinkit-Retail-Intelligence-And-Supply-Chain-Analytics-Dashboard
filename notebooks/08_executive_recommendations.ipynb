```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
```

```python
customer_segments = pd.read_csv(
    "../exports/segmentation_output/customer_segments.csv"
)

clv = pd.read_csv(
    "../exports/processed_data/customer_clv.csv"
)

inventory_master = pd.read_csv(
    "../exports/supply_chain_output/inventory_master.csv"
)

category_inventory_health = pd.read_csv(
    "../exports/supply_chain_output/category_inventory_health.csv"
)

marketing_summary = pd.read_csv(
    "../exports/processed_data/marketing_summary.csv"
)

kpis = pd.read_csv(
    "../exports/processed_data/master_kpis.csv"
)

orders = pd.read_csv(
    "../exports/cleaned_data/orders_clean.csv"
)

customers = pd.read_csv(
    "../exports/cleaned_data/customers_clean.csv"
)

feedback = pd.read_csv(
    "../exports/cleaned_data/feedback_clean.csv"
)

marketing = pd.read_csv(
    "../exports/cleaned_data/marketing_clean.csv"
)
```

```python
executive_kpis = {
    "Total Revenue":
        orders["order_total"].sum(),

    "Total Orders":
        len(orders),

    "Total Customers":
        customers["customer_id"].nunique(),

    "Average Order Value":
        orders["order_total"].mean(),

    "Average Customer CLV":
        clv["clv"].mean(),

    "High Risk Products":
        len(
            inventory_master[
                inventory_master["risk_category"]
                == "High Risk"
            ]
        ),

    "Average Rating":
        feedback["rating"].mean(),

    "Average ROAS":
        marketing["roas"].mean()
}
```

```python
executive_summary_df = pd.DataFrame(
    executive_kpis.items(),
    columns=[
        "Metric",
        "Value"
    ]
)

executive_summary_df
```

```python
high_value_customers = clv[
    clv["clv_segment"]
    ==
    "Very High Value"
]

print(f"""
High Value Customers:
{len(high_value_customers)}

Recommendation:
Retain these customers using loyalty rewards,
exclusive discounts and personalized offers.
""")
```

```python
lowest_health_categories = (
    category_inventory_health
    .sort_values(
        "inventory_health_score"
    )
    .head(5)
)

lowest_health_categories
```

```python
print("""
Inventory Recommendation

Focus inventory optimization efforts on
low-health categories.

Priority Categories:
• Household Care
• Dairy & Breakfast
• Fruits & Vegetables
""")
```

```python
risk_summary = (
    inventory_master[
        "risk_category"
    ]
    .value_counts()
)

risk_summary
```

```python
print("""
Supply Chain Recommendation

High-risk products should be reviewed
for replenishment and damage reduction.

Priority:
Reduce damaged stock and improve
inventory monitoring.
""")
```

```python
marketing_summary.sort_values(
    "roas",
    ascending=False
)
```

```python
best_channel = (
    marketing_summary
    .sort_values(
        "roas",
        ascending=False
    )
    .iloc[0]
)

print(f"""
Marketing Recommendation

Top Performing Channel:
{best_channel['channel']}

Recommendation:
Increase investment in high ROAS channels.
""")
```
